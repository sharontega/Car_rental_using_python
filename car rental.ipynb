The aim of the project is to analyze and answer some car rental questions in different states in united state.

The dataset was obtained from kaggle.

```python
#importing important modules
import pandas as pd
import numpy as py
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#importing the dataset
df=pd.read_csv('CarRentalDataV1.csv')
df.head()
```


# EDA

```python
#looking at the columns in the dataset
df.columns
```

```python
#getting the shape of the data
df.shape
```

```python
df.info()
```

```python
#removing duplicate records
df=df.drop_duplicates()
```

```python
df.shape
```

```python
#separating the numerical columns into numerical and categorical values to have a look at the distribution of the data
num=df[['rating', 'renterTripsTaken', 'reviewCount','owner.id', 'rate.daily','vehicle.year']]
cat=df[['fuelType','location.city', 'location.country','location.state','vehicle.make', 'vehicle.model', 'vehicle.type','airportcity']]
```

```python
#plotting the numerical values
for i in num:
    sns.histplot(num[i])
    plt.title(i)
    plt.show()
```

```python
#plotting the categorica values
for i in cat:
    sns.countplot(cat[i])
    plt.title(i)
    plt.show()
```

```python
#checking for null values
df.isnull().sum()
```

```python
#removing location from column name
df=df.rename(columns=lambda x: x.replace('location.',''))
```

```python
#removing vechikle from column name
df=df.rename(columns=lambda x:x.replace('vehicle.',''))
```

```python
#replacing string in column name
df=df.rename(columns=lambda x:x.replace('.','_'))
```

```python
#changing year datatype from float to int
df['year']=df.year.astype(int)
```

```python
#dropping nan values in fueltype
df=df.dropna(subset=['fuelType'])
```

```python
#filling nan values with zero in rating
df['rating']=df.rating.fillna(0)
```

```python
#dropping unwanted columns
df=df.drop(['latitude','longitude','country'],axis='columns')
```

# visualization

```python
#1 what fueltype is mostly used?
sns.countplot(df.fuelType)
```

```python
#2 city with the highest rental
plt.figure(figsize=(20,6))
sns.countplot(df.city,order=df.city.value_counts().iloc[:10].index)
```

```python
#3 owners with the higest rental
plt.figure(figsize=(20,6))
sns.countplot(df.owner_id,order=df.owner_id.value_counts().iloc[:10].index)
```

```python
#4 most popular make in each cities
df.groupby('city')['make'].apply(pd.Series.mode).to_frame().head(10)
```

```python
#5 most popular model in each cities
df.groupby('city')['model'].apply(pd.Series.mode).to_frame().head(10)
```
